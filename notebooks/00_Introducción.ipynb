# Predicción de Volatilidad y Pricing de Opciones: GARCH vs Deep Learning (LSTM & Transformers) vs XGBoost

# 1. Introducción y Contexto

En las finanzas cuantitativas, la volatilidad no es simplemente una medida de riesgo, sino una variable fundamental para la valoración de instrumentos derivados. Un derivado es un contrato financiero cuyo valor "deriva" del precio de otro activo, denominado activo subyacente (como el SPY). A diferencia de los precios, la volatilidad no es observable directamente en el mercado, lo que convierte su estimación en uno de los mayores retos del sector.

Este proyecto compara la eficacia de modelos estadísticos tradicionales, algoritmos de Machine Learning y arquitecturas de Deep Learning para predecir la volatilidad del S&P 500 (datos de 2000 a 2026) y evaluar su impacto en el Pricing de Opciones.

# 2. ¿Qué es el Volatility Forecasting?

Es el proceso de predecir la variabilidad futura de los rendimientos. Es crucial porque la volatilidad presenta "hechos estilizados" (patrones estadísticos recurrentes):

- Volatility Clustering: Los periodos de alta volatilidad tienden a ser seguidos por alta volatilidad, y los de baja por baja.

- Mean Reversion: La volatilidad suele regresar a una media histórica tras eventos extremos.

- Leverage Effect: Existe una correlación negativa entre los retornos y la volatilidad; las caídas de precio generan aumentos de volatilidad mucho más agresivos que las subidas.

# 3. Metodologías Utilizadas

## 3.1. GARCH (Generalized Autoregressive Conditional Heteroskedasticity)

El modelo GARCH es el estándar de la econometría clásica. A diferencia de una media móvil simple, el GARCH asume que la varianza actual está en función de los errores pasados y de las varianzas pasadas. Su función es capturar la persistencia de la volatilidad y el efecto de agrupamiento. La ecuación básica GARCH(1,1) es:
$$\sigma_t^2 = \omega + \alpha \epsilon_{t-1}^2 + \beta \sigma_{t-1}^2$$
Donde $\sigma_t^2$ es la varianza prevista, $\epsilon_{t-1}^2$ es el residuo del periodo anterior y $\sigma_{t-1}^2$ es la varianza anterior.

## 3.2. LSTM (Long Short-Term Memory)

Las LSTM son un tipo de Red Neuronal Recurrente (RNN) diseñadas específicamente para aprender dependencias a largo plazo en series temporales.\
A diferencia de los modelos lineales, las LSTM pueden capturar relaciones no lineales complejas en los datos históricos. Gracias a sus "puertas" (gates) de entrada, salida y olvido, deciden qué información del pasado es relevante mantener para predecir la volatilidad futura, mitigando el problema del desvanecimiento del gradiente (la tendencia de la red a olvidar patrones de largo plazo durante el entrenamiento).

## 3.3. Transformers y Mecanismos de Atención

A diferencia de las LSTM que procesan los datos secuencialmente, los Transformers pueden observar toda la serie temporal a la vez. El mecanismo de atención asigna "pesos" a diferentes momentos del pasado, identificando qué eventos históricos (por ejemplo, la crisis de 2008 o la de 2020) tienen más relevancia para la predicción actual, sin importar cuán lejos estén en el tiempo.

## 3.4. XGBoost (Extreme Gradient Boosting)

Es un algoritmo de aprendizaje supervisado basado en un ensamble de árboles de decisión. A diferencia de las redes neuronales, XGBoost construye árboles de forma secuencial donde cada nuevo árbol intenta corregir los errores de los anteriores (Gradient Boosting). Es extremadamente eficiente con datos tabulares y permite identificar la importancia de las variables (Feature Importance), revelando qué "lags" o indicadores técnicos tienen más peso en el movimiento de la volatilidad.

# 4. Aplicación: Pricing de Opciones

El objetivo final de predecir la volatilidad ($\sigma$) es utilizarla como input en modelos de valoración de opciones, principalmente en el modelo de **Black-Scholes-Merton**.\
La fórmula de Black-Scholes para una opción Call requiere cinco variables:
1. Precio del activo ($S$)
2. Precio de ejercicio ($K$)
3. Tiempo hasta el vencimiento ($T$)
4. Tasa libre de riesgo ($r$)
5. Volatilidad ($\sigma$)

Dado que $\sigma$ es la única variable no observable y la más sensible, una predicción más precisa de la volatilidad se traduce directamente en un cálculo del "valor justo" (fair value) de la opción más exacto, permitiendo identificar oportunidades de arbitraje o gestionar mejor el riesgo de una cartera.

# 5. Datos y Herramientas

- Dataset: Datos históricos del S&P 500 (^GSPC / SPY) vía yfinance.

- Periodo: 2000 - 2026 (incluyendo la burbuja Dot-com, la crisis de 2008, la pandemia de 2020 y el régimen actual).

- Librerías: arch (GARCH), TensorFlow/Keras (Deep Learning), XGBoost y scipy (Black-Scholes).

# 6. Objetivos del Proyecto

1. Benchmark: Comparar si la complejidad de los Transformers y LSTMs realmente supera la capacidad predictiva de un modelo GARCH tradicional en diferentes regímenes de mercado.

2. Análisis de Error: Evaluar el rendimiento usando métricas como RMSE (Root Mean Squared Error) y MAE (Mean Absolute Error).

3. Implementación de Pricing: Desarrollar un módulo que tome la volatilidad predicha y calcule el precio de opciones teóricas, comparándolas con los precios implícitos de mercado.
